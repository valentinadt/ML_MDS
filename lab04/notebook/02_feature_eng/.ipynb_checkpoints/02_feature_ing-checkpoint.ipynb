At first we imported the entire data set provided by MovieLens which includes 25.000.095 ratings and 1.093.360 tag applications across 62.423 movies. Later on, we will work only with the most relevant ones for the analysis.

```python
df_scores = pd.read_csv('../../data/genome-scores.csv', sep= ",")
df_g_tags = pd.read_csv('../../data/genome-tags.csv', sep= ",")
df_movies = pd.read_csv('../../data/movies.csv', sep= ",")
df_ratings = pd.read_csv('../../data/ratings.csv', sep= ",")
df_links = pd.read_csv('../../data/links.csv', sep= ",")
df_tags = pd.read_csv("../../data/tags.csv", sep =",")
```

First, we will see how the different datasets are organised.

```python
# Merge the movies and ratings datasets by movieId

data_movies = pd.merge(df_movies, df_ratings, on='movieId', how='inner')

# Visualization

data_movies.head()
```

#### Movies Dataset

```python

# Creating a new column called 'year' by using  the year values from the 'title' column

df_movies['year'] = df_movies.title.str.extract('\s\((\d+)', expand=True)
# quitar los años a la columna titulo
df_movies['title'] = df_movies['title'].str.replace(r"\(.*\)","")
df_movies.head()


# https://stackoverflow.com/questions/51866182/how-do-i-split-multiple-rows-in-different-column-in-pandas
```

#### Merged Dataset 

```python

# Creating a new column called 'year' by using  the year values from the 'title' column

data_movies['year'] = data_movies.title.str.extract('\s\((\d+)', expand=True)
# quitar los años a la columna titulo
data_movies['title'] = data_movies['title'].str.replace(r"\(.*\)","")
#data_movies.head()
```

```python
df_movies.head() 
```

```python
# Getting series of lists by applying split operation.
df_movies.genres = df_movies.genres.str.split('|')
# Getting distinct genre types for generating columns of genre type.
genre_columns = list(set([j for i in df_movies['genres'].tolist() for j in i]))
# Iterating over every list to create and fill values into columns.
for j in genre_columns:
    df_movies[j] = 0
for i in range(df_movies.shape[0]):
    for j in genre_columns:
        if(j in df_movies['genres'].iloc[i]):
            df_movies.loc[i,j] = 1

```

```python
df_movies.head()
```

```python
# Dropping 'genre' columns as it has already been one hot encoded.

df_movies.drop('genres',axis=1,inplace=True)
df_movies.head()
```

```python
data_movies = pd.merge(df_movies, df_ratings, on='movieId', how='inner')
data_movies.head()
```

```python
data_movies.drop('IMAX',axis=1,inplace=True)
```
